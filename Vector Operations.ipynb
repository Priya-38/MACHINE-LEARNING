```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

##### Plotting Vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,5,5, scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,5,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,5,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

###### Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,5,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.quiver(0,0,5,-1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

###### Multiplying a vector by a scaler

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
product = vector_1 * vector_2
print(product)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = -0.5 * vector_2
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

###### Dot product of two vectors

```python
a = np.array([2,3])
b = np.array([4,5])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

```python
a = np.array([2,3,8])
b = np.array([4,5,9])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

###### Cross product of two vectors

```python
a = np.array([2,3])
b = np.array([4,5])
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

```python
a = np.array([2,3,8])
b = np.array([4,5,9])
a_cross_b = np.cross(a,b)
```
